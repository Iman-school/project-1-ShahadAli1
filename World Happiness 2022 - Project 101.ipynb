```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
World_Happiness_Report_2022 = pd. read_csv( 'World Happiness Report 2022.csv') 
World_Happiness_Report_2022
```

```python
x= World_Happiness_Report_2022["Happiness score"] 
y= World_Happiness_Report_2022["Explained by: GDP per capita"] 
plt.scatter (x, y)
plt.xlabel("Happiness score")
plt.ylabel("Explained by: GDP per capita")
```

```python
from sklearn. linear_model import LinearRegression
x= World_Happiness_Report_2022["Happiness score"].values.reshape (-1,1)
y = World_Happiness_Report_2022["Explained by: GDP per capita"]
lr_model = LinearRegression() 
lr_model.fit(x, y) 
y_pred = lr_model.predict(x)
y_pred
                
plt.scatter(x, y)
plt.xlabel("Happiness score") 
plt.ylabel("Explained by: GDP per capita")
plt.plot(x, y_pred)
```

```python
theta_0 = lr_model.intercept_
theta_1= lr_model.coef_
theta_0, theta_1
```
